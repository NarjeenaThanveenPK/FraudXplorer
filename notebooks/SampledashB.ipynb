```python
import pandas as pd

# Load the remaining rows
remaining_df = pd.read_csv('paysim.csv', skiprows=500001, header=None)

# Set the correct column names
remaining_df.columns = ['step', 'type', 'amount', 'nameOrig', 'oldbalanceOrg',
                        'newbalanceOrig', 'nameDest', 'oldbalanceDest',
                        'newbalanceDest', 'isFraud', 'isFlaggedFraud']

# Encode categorical
remaining_df['type'] = remaining_df['type'].astype('category').cat.codes

# Drop unwanted columns
remaining_df.drop(['nameOrig', 'nameDest'], axis=1, inplace=True)

# Ensure both classes are included
fraud = remaining_df[remaining_df['isFraud'] == 1].sample(50, random_state=1)
non_fraud = remaining_df[remaining_df['isFraud'] == 0].sample(60, random_state=1)

# Combine and shuffle
dashboard_df = pd.concat([fraud, non_fraud]).sample(frac=1, random_state=1)

# Save
dashboard_df.to_csv("dashboard_sample1.csv", index=False)
print("✅ dashboard_sample1.csv created!")
```
